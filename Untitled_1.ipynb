```python
import numpy as np
import pandas as pd
from itertools import zip_longest
import plotly.express as px
from datetime import datetime, time
import re
from collections import Counter
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('test.csv', encoding='latin1')
df
```

```python
df[['area', 'governorate', 'country']] = df['location'].str.split(',', expand=True)
df
```

```python
df.columns
```

```python
df.columns=df.columns.str.lower().str.strip().str.replace(' ','_')
df['company_name'] = df['company_name'].str.replace('-','')
df
```

```python
df['Experience'] = df['skills'].apply(lambda x: re.search(r'\d+ - \d+ Yrs of Exp|\d+\+ Yrs of Exp', x).group() if re.search(r'\d+ - \d+ Yrs of Exp|\d+\+ Yrs of Exp', x) else np.NAN)
df
```

```python
split_columns = df['skills'].str.split(' · ', expand=True)
split_columns
```

```python
df['min_experience'] = df['Experience'].str.extract(r'(\d+)').astype(float)
df['max_experience'] = df['Experience'].str.extract(r'(\d+) - (\d+)')[1].astype(float)
df['max_experience'] = df['max_experience'].fillna(df['min_experience'])
df['min_experience'] = df['min_experience'].astype(int)
df['max_experience'] = df['max_experience'].astype(int)
df
```

```python
df['Average_Experience'] = (df['min_experience'] + df['max_experience']) / 2
df.to_csv('test2.csv', index=False)
df
```

Most required **skills**

```python
split_columns = df['skills'].str.split(' Â· ', expand=True)
split_columns = split_columns.iloc[:, 3:]
split_columns = split_columns.replace(['Cost Accounting (1.0)','General ledger accountant','General Ledger','General Accounting','accountant','Accounting', 'Accounting Software','Accounting principles','Accounting/Finance','Accounts Payable','Accounts Receivable','general accountant'], 'Accountant')
split_columns = split_columns.replace(['Microsoft Office','MS ','excel','pivot','Microsoft Excel','MS Excel','MS Office','Vlookups',], 'office')
split_columns = split_columns.replace(['ERP','Oracle','SAP ','ERP System','QuickBooks ','Quick books','Zoho '], 'ERP ')
split_columns = split_columns.replace(['Financial Accountant','Financial', 'Finance', 'Financial Accounting', 'Financial Analysis', 'Financial Management', 'Financial Reporting'], 'Finance')
split_columns = split_columns.replace(['Audit', 'GAAP', 'IFRS'], 'Audit & Standards')
split_columns = split_columns.replace(
    ['Treasury', 'Treasury Accountant', 'Tax'], 'Treasury & Tax')
split_columns = split_columns.replace(
    ['Ledger', 'Reporting', 'reporting'], 'Ledger & Reporting')
split_columns = split_columns.replace(
    ['Business Administration', 'business'], 'Business')

split_columns = split_columns.replace(
    ['Management', 'Financial Management', 'Operations'], 'Management')

split_columns = split_columns.replace(
    ['Communication', 'Communication skills'], 'Communication')
split_columns = split_columns.replace(
    ['analyst', 'analytical'], 'Analytical Skills')
split_columns = split_columns.replace(
    ['Manufacturing', 'Manufacturing/Production', 'Sheet Metal'], 'Manufacturing')

skill_counts = split_columns.apply(pd.Series.value_counts).sum(axis=1)
skill_counts = skill_counts.sort_values(ascending=False)
skill_counts.to_csv('skill_counts.csv', index=True)
skill_counts
```
